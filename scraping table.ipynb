```python
#Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

url='https://ticker.finology.in/'
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text,'lxml')
soup
```

```python
table=soup.find('table',class_='table table-sm table-hover screenertable')
```

```python
table
```

```python
#extract title
titles=[]
headers=table.find_all('th')
for i in headers:
    title=i.text
    titles.append(title)
titles
```

```python
#get rows
datalist=[]
rows=table.find_all('tr')
for row in rows[1::]:
    data=row.find_all('td')
    li=[i.text for i in data]
    datalist.append(li)
datalist
```

```python
#create dataframe
df=pd.DataFrame(datalist,columns=titles)
df
```

```python
#remove \n from company
df['Company']=df['Company'].str.replace('\n','')
df
```
